```python
import pandas as pd
```

```python
df = pd.DataFrame({'Sizes':['s','m','l','xl','xxl','xxs','s','m','l','xl','xxl','xxs']})
```

```python
df
```

```python
ord_df = [['s','m','l','xl','xxl','xxs']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_df)
df['Encoded'] = oe.fit_transform(df)
```

```python
df
```

**Mapping**

```python
df1 = pd.DataFrame({'Size':['s','m','l','xl','xxl','xxs','s','m','l','xl','xxl','xxs']})
```

```python
ord_data = {'s':5,'m':4,'l':3,'xl':2,'xxl':1,'xxs':0}
ord_data
```

```python
df1['Size_en_map'] = df1['Size'].map(ord_data)
```
