```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = [np.random.randint(10,100) for i in range(10000)]
```

```python
len(data)
```

```python
pop_table = pd.DataFrame({"pop_data":data})
pop_table.head()
```

```python
np.mean(data)
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(pop_table)
plt.show()
```

```python
sample_mean = []
for no_sample in range(60):
  sample_data = []
  for i in range(500):
    sample_data.append(np.random.choice(data))
  sample_mean.append(np.mean(sample_data))
```

```python
np.mean(sample_mean)
```

```python
sample_Mean = pd.DataFrame({"sample_mean":sample_mean})
sample_Mean.head()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="sample_mean", data=sample_Mean)
plt.show()
```
